```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

#Lendo o arquivo
M= Image.open ('Architecture.jpg')
#Convertendo em matriz
M= np.asarray(M, dtype=np.float32)/255
#Visualização
plt.imshow(M, aspect="auto")
plt.axis("off")
plt.show()

M.shape

M[:,:,1] = 0.3 * M[:,:,1] # diminuir R em 30%
M[:,:,2] = 0.6 * M[:,:,2] #diminuir G em 60%
M[:,:,0] = 1 * M[:,:,0]

import imageio
M= 255 * M
M= M.astype(np.uint8)
imageio. imwrite('architecture_yellow_filter.jpg', M)
```
